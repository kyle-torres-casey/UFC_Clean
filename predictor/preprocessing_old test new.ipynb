```python
import pandas as pd
import numpy as np
```

```python
# read data for fights and fighters
fighters = pd.read_csv("all_fighters.csv")
```

Next, We will perform a series of steps to clean up the datasets.

```python
fighters.head()
```

```python
fighters['Name'] = fighters['First'].fillna('') + ' ' + fighters['Last'].fillna('')

# Strip any leading or trailing whitespace (in case both First and Last are NaN)
fighters['Name'] = fighters['Name'].str.strip()

# # check if there are fighters with the same name
# fighters[fighters.duplicated(subset="Name", keep=False)]
```

```python
fighters[fighters.duplicated(subset="Name", keep=False)]
```

```python
columns = ['Name'] + [col for col in fighters.columns if col != 'Name']
fighters = fighters[columns]
fighters.head()
```

```python
fighters.drop(3325, inplace=True)
fighters.iloc[3311, 0] = "Bruno Silva 185"
fighters.drop(2627, inplace=True)
fighters.drop(2103, inplace=True)
fighters.iloc[1988, 0] = "Tony Johnson 265"
fighters.iloc[480, 0] = "Michael McDonald 205"
```

```python
columns = ['Name'] + [col for col in fighters.columns if col != 'Name']
fighters = fighters[columns]
fighters.head()
```

```python
# use fighter names as index
# fighters.drop("fighter_id", axis=1, inplace=True)
fighters.set_index("Name", inplace=True)
```

```python
# Some fighters do not have statistics available, and we will remove those fighters.
fighters_clean = fighters.loc[~((fighters["SLpM"] == 0) &
                               (fighters["Str. Acc."] == "0%") &
                               (fighters["SApM"] == 0) &
                               (fighters["Str. Def"] == "0%") &
                               (fighters["TD Avg."] == 0) &
                               (fighters["TD Acc."] == "0%") &
                               (fighters["TD Def."] == "0%") &
                               (fighters["Sub. Avg."] == 0))].copy()
```

```python
print("{0} fighters in total, after clean up: {1} fighters".format(len(fighters), len(fighters_clean)))
```

Note:
* SLpM - Significant Strikes Landed per Minute
* Str_Acc - Significant Striking Accuracy
* SApM - Significant Strikes Absorbed per Minute
* Str_Def - Significant Strike Defence (the % of opponents strikes that did not land)
* TD_Avg - Average Takedowns Landed per 15 minutes
* TD_Acc - Takedown Accuracy
* TD_Def - Takedown Defense (the % of opponents TD attempts that did not land)
* Sub_Avg - Average Submissions Attempted per 15 minutes

```python
fighters_clean["win%"] = 100 * (fighters_clean["W"] / (fighters_clean["W"] +
                                                       fighters_clean["L"] +
                                                       fighters_clean["D"]))
                                                      #  fighters_clean["nc"]))  # Assuming nc is not in use

# Columns that have percentages as strings
percentages = ["Str. Acc.", "Str. Def", "TD Acc.", "TD Def."]
# Columns to convert to float32
statistics = ["SLpM", "Str. Acc.", "SApM", "Str. Def", "TD Avg.", "TD Acc.", "TD Def.", "Sub. Avg.", "win%"]

# Remove the '%' sign and convert to float
for col in percentages:
    fighters_clean[col] = fighters_clean[col].map(lambda x: float(x.replace("%", "")) if pd.notna(x) else x)

# Convert statistics columns to float32
fighters_clean[statistics] = fighters_clean[statistics].astype(np.float32, errors='ignore')

print(fighters_clean)
```

```python
fighters_clean.head()
```

```python
# use born year only
def get_year(dob):
    if pd.isna(dob):
        return "unk"
    return dob.split(", ")[-1]

fighters_clean["born_year"] = fighters_clean["DOB"].map(lambda dob: get_year(dob))
```

```python
fighters_clean.head()
```

```python
# remove unused columns
fighters_clean.drop(["DOB"], inplace=True, axis=1)
fighters_clean.drop(["Unnamed: 15", "First", "Last", "Nickname"], inplace=True, axis=1)
fighters_clean.drop(["Wt"], inplace=True, axis=1)
fighters_clean.drop(["Belt"], inplace=True, axis=1)
```
