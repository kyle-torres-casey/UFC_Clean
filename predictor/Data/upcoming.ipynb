```python
import pandas as pd
bouts = pd.read_csv("bouts_upcoming_0927_2.csv")
```

```python
bouts.columns
```

```python
event = bouts[(bouts['Event'] == 'UFC 307: Pereira vs. Rountree Jr.')]
```

```python
# event
```

```python
import numpy as np

# Assuming your DataFrame is called bouts_clean
fighter_1_unique = event['Fighter 1'].unique()
fighter_2_unique = event['Fighter 2'].unique()

# Combine unique values from both columns
all_unique_fighters = np.union1d(fighter_1_unique, fighter_2_unique)

# Print or inspect the unique fighters
print(all_unique_fighters)
```

```python
# Create a new DataFrame with a column called 'Name'
unique_fighters_df = pd.DataFrame(all_unique_fighters, columns=['Name'])
```

```python
unique_fighters_df.to_csv("names_307.csv")
```

```python
# Reset index in place (modifies the original DataFrame)
event.reset_index(drop=True, inplace=True)
```

```python
# event
```

```python
import pandas as pd

# Iterate through each row with index
rows_to_drop = set()
for idx1, row1 in event.iterrows():
    # Check all subsequent rows
    for idx2, row2 in event.iterrows():
        if idx1 != idx2 and idx2 not in rows_to_drop and idx1 not in rows_to_drop:
            # Check if "Fighter 1", "Fighter 2", and "Date" are the same
            if (
                (row1['Fighter 1'] == row2['Fighter 1'] and
                row1['Fighter 2'] == row2['Fighter 2'] and
                row1['Date'] == row2['Date']) | 
                (row1['Fighter 2'] == row2['Fighter 1'] and
                row1['Fighter 1'] == row2['Fighter 2'] and
                row1['Date'] == row2['Date'])
            ):
                # Mark the second row for dropping
                print(idx2)
                rows_to_drop.add(idx2)

# Drop the matching rows
bouts_cleaned = event.drop(index=list(rows_to_drop))

# Inspect cleaned dataframe
print(bouts_cleaned)
```
