```python
import pandas as pd
df1 = pd.read_csv('combined_fight_odds_916.csv')
df2 = pd.read_csv('fights_odds_paris_fighters.csv')
# missing_fights_916
# Concatenate the DataFrames
# combined_df = pd.concat([df1, df2, df3, df4, df5, df6])
combined_df = pd.concat([df1, df2])

# Drop duplicate rows
combined_df = combined_df.drop_duplicates()

# Reset the index
combined_df = combined_df.reset_index(drop=True)

# Display the final DataFrame
print(len(combined_df))
# df.head()
```

```python
combined_df.head()
```

```python
# Sort rows to ensure consistent ordering
df_sorted = combined_df.sort_values(by=['Date', 'Fighter', 'Opponent']).reset_index(drop=True)

# Create a boolean mask to identify duplicates based on the condition
duplicates_mask = (
    (df_sorted['Fighter'] == df_sorted['Opponent'].shift()) &
    (df_sorted['Opponent'] == df_sorted['Fighter'].shift()) &
    (df_sorted['Date'] == df_sorted['Date'].shift())
)

# Filter the DataFrame to remove duplicates
df_no_duplicates = df_sorted[~duplicates_mask].reset_index(drop=True)

# Display the DataFrame without duplicates
print(df_no_duplicates)
print(len(df_no_duplicates))
```

```python
df = df_no_duplicates
```

```python
# df = df_no_duplicates
# Combine the 'Fighter' and 'Opponent' columns
combined_names = pd.concat([df['Fighter'], df['Opponent']])

# Get unique names
unique_names = combined_names.unique()

# Convert to a list (optional)
unique_names_list = list(unique_names)
```

```python
import pandas as pd

# Iterate through each row with index
rows_to_drop = set()
for idx1, row1 in df2.iterrows():
    # Check all subsequent rows
    for idx2, row2 in df2.iterrows():
        if idx1 != idx2 and idx2 not in rows_to_drop and idx1 not in rows_to_drop:
            # Check if "Fighter 1", "Fighter 2", and "Date" are the same
            if (
                (row1['Fighter'] == row2['Fighter'] and
                row1['Opponent'] == row2['Opponent'] and
                row1['Date'] == row2['Date']) | 
                (row1['Opponent'] == row2['Fighter'] and
                row1['Fighter'] == row2['Opponent'] and
                row1['Date'] == row2['Date'])
            ):
                # Mark the second row for dropping
                print(idx2)
                rows_to_drop.add(idx2)

# Drop the matching rows
odds_cleaned = df2.drop(index=list(rows_to_drop))

# Inspect cleaned dataframe
print(odds_cleaned)
```
